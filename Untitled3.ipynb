Attribute Information:
About client features:

- age

- job (type of job)

- marital (marital status)

- education

- default (has credit in default? 'no','yes','unknown')

- balance

- housing (has housing loan?'no','yes','unknown')

- loan (has personal loan?'no','yes','unknown')

About current campaign:
- contact (communication type: 'cellular','telephone')

- duration (last contact duration, in seconds) Important note: the duration attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.

- day (last contact day of the week)

- month (last contact month of year)

- campaign (number of contacts performed during this campaign and for this client)

- deposit (has the client subscribed a term deposit?'yes','no')

About previous contacts/campaign:

- pdays (number of days that passed by after the client was last contacted from a previous campaign, 999 means client was not previously contacted)

- previous (number of contacts performed before this campaign and for this client)

- poutcome (outcome of the previous marketing campaign: 'failure','nonexistent','success')

Import Library

```
# Import Library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
%matplotlib

```

Import Raw Dataset

```
# Import Data to Google Colab
df_churn = pd.read_csv ('https://raw.githubusercontent.com/Khayrussyifapr/Bank-Marketing-Classification/main/bank.csv', sep = ',')
df_bank = df_churn

# Show 10 First Row
df_bank.head()
```

```
# Prints the Dataset Information
df_bank.info()
```

```
# Prints Descriptive Statistics
df_bank.describe().transpose()
```

```
# Check for Missing Values
df_bank.isnull().sum()
```

Pre-Processing

Merubah target nilai kelas dari object ke numerical /
dari "Yes-No" menjadi "1-0"

```
encoder = LabelEncoder()
df_bank['deposit'] = encoder.fit_transform(df_bank.deposit.values)

df_bank = pd.get_dummies(df_bank)
```

```
df_bank.head()
```

```
df_bank.deposit
```

```
feature = df_bank.drop('deposit', axis =1) 
target = df_bank.deposit
feature
```

```
target
```

Selanjutnya data dibagi menjadi data train dan data set. Data train dilakukan sebagai pembelajaran atar pelatihan agar ketika diuji atau diprediksi model dapat memberikan hasil yang terbaik. Semakin tinggi nilai data test maka semakin baik hasil pemodelan. Apabila nilai hasil data train dan data set diatas 70% maka pemodelan sudah baik. 

```
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.30, random_state=42)
```

Pemodelan Decision Tree

```
tree_model = DecisionTreeClassifier(max_depth=8).fit(X_train, y_train)
```

```
accuracy_train = tree_model.score(X_train, y_train)
accuracy_test = tree_model.score(X_test, y_test)
print("Akurasi pada training set: ",round(accuracy_train*100,2),'%')
print("Akurasi pada test set: ",round(accuracy_train*100,2),'%')

```

Hasil akurasi data training dan data set menggunakan pemodelan decision tree menunjukan nilai yang sama persentasenya dan melebihi 70% sehingga dapat dikatakan bahwa nilai pemodelan baik atau bagus

```
y_preds = tree_model.predict(X_test)
print("Confusion Matrix :\n", confusion_matrix(y_preds,y_test), "\n\n")
print("Classification Report :\n", classification_report(y_preds,y_test))
```

Pemodelan KNN 

```
knn_model = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)
```

```
accuracy_train_knn = knn_model.score(X_train, y_train)
accuracy_test_knn = knn_model.score(X_test, y_test)
print("Akurasi pada Training Set: ", round(accuracy_train_knn*100, 2), '%')
print("Akurasi pada Test Set: ", round(accuracy_test_knn*100, 2), '%')
```

```
y_preds_knn = knn_model.predict(X_test)
print("Confusion Matrix : \n", confusion_matrix(y_preds_knn, y_test), "\n\n")
print("Classification Report : \n", classification_report(y_preds_knn, y_test))
```

Hasil Pemodelan KNN
Akurasi pada Training Set:  84.95 %
Akurasi pada Test Set:  73.63 %

Walaupun hasil pemodelan memberikan nilai diatas 70% tetapi terdapat perbedaan yang lumayan besar antara data train dan data test, sehingga apabila ingin dilakukan perbandingan kedua model antara decision tree dan KKN maka dapat dilihat bahwa pemodelan decision tree lebih unggul karena nilai antara data train dan data set diatas 70% dan seimbang.
